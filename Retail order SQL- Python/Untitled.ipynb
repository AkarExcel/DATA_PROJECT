```python
# extract the file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#read data from the file and handle null values
import pandas as pd
# fix the incorrect values in the Ship Mode Field
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
# Check for unique values in Ship Mode Field
df['Ship Mode'].unique()
```

```python
# rename column names that are not using the best practice
#convert name to lowercase  
df.columns = df.columns.str.lower()
# replace space(" ") with underscore"_"
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Derive column for Discount, sales price and profit

#discount price
df['discount'] = df['list_price']*df['discount_percent']*0.01

#sales price
df['sale_price'] = df['list_price'] - df['discount']

#profit

df['profit'] = df['sale_price'] - df['cost_price']
```

```python
# check datatypes
df.dtypes

#convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
# remove unused column
df.drop(columns=['list_price','cost_price','sales_price','discount_percent'], inplace = True)
```

```python
# load the data into sql server using replace option
import sqlalchemy as sql
engine = sql.create_engine('mssql://E3/pysql?driver=SQL+SERVER+NATIVE+CLIENT+11.0')
conn = engine.connect()
```

```python
#load the data into sql server using the append option
df.to_sql('df_orders',con=conn, index=False, if_exists = 'append')
```
